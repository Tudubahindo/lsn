### Overview 

<span style="color:gray">In this notebook our task will be to perform machine learning regression on noisy data with a Neural Network (NN).

<span style="color:gray">We will explore how the ability to fit depends on the structure of the NN. The goal is also to build intuition about why prediction is difficult.

### The Prediction Problem

<span style="color:gray">Consider a probabilistic process that gives rise to labeled data $(x,y)$. The data is generated by drawing samples from the equation

$$
    y_i= f(x_i) + \eta_i,
$$

<span style="color:gray">where $f(x_i)$ is some fixed, but (possibly unknown) function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that

$$
\langle \eta_i \rangle=0 \\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$

<span style="color:gray">We will refer to the $f(x_i)$ as the **true features** used to generate the data. 

<span style="color:gray">To make predictions, we will consider a NN that depends on its parameters, weights and biases. The functions that the NN can model respresent the **model class** that we are using to try to model the data and make predictions.

<span style="color:gray">To learn the parameters of the NN, we will train our models on a **training data set** and then test the effectiveness of the NN on a *different* dataset, the **validation data set**. The reason we must divide our data into a training and test dataset is that the point of machine learning is to make accurate predictions about new data we have not seen.

<span style="color:gray">To measure our ability to predict, we will learn our parameters by fitting our training dataset and then making predictions on our test data set. One common measure of predictive  performance of our algorithm is to compare the predictions,$\{y_j^\mathrm{pred}\}$, to the true values $\{y_j\}$. A commonly employed measure for this is the sum of the mean square-error (MSE) on the test set:</span>
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$

<span style="color:gray">We will try to get a qualitative picture by examining plots on validation and training data.

### Linear fit

<span style="color:gray">We start by considering the very simple case:</span>
$$
f(x)=2x+1
$$

<span style="color:gray">Let's start defining the parameters of an ideal linear function which we are going to predict through a neural network regression

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

<span style="color:gray">Now let's generate a set of input data which will slightly deviate from our ideal behaviour using a random noise (that actually is set to zero):

```python
import numpy as np
import os
import random

os.environ['PYTHONHASHSEED']=str(0)
random.seed(0)
np.random.seed(0)

# generate training inputs
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

<span style="color:gray">If you remember how a single node of a neural network works, you can easily spot that **just a single neuron can make the job**. So let's start using a simple Sequential model with just one layer on one neuron only!

```python
# compose the NN model
import tensorflow as tf
tf.random.set_seed(0)
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

#session_conf = tf.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)
#sess = tf.Session(graph=tf.get_default_graph(), config=session_conf)
#K.set_session(sess)

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

<span style="color:gray">We are now going to train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function (out linear regression function).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

<span style="color:gray">By looking at weights and biases we can see if the linear fit was successfull: $w_1$ represents the angular coefficient, $b$ the intercept.

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

### Exercise 11.2

<span style="color:gray">Try to extend the model to obtain a reasonable fit of the following polynomial of order 3:

$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.

<span style="color:gray">Find good (& reasonable) choices for:

- the number of layers
- the number of neurons in each layer
- the activation function
- the optimizer
- the loss function

<span style="color:gray">Check your NN model by seeing how well your fits predict newly generated test data (including on data outside the range you fit. How well do your NN do on points in the range of $x$ where you trained the model? How about points outside the original training data set? 
Summarize what you have learned about the relationship between model complexity (number of parameters), goodness of fit on training data, and the ability to predict well.

### Solution:

Considering the complexity of the target function, I decided to use a training dataset of 5000 points (and a validation dataset of 500 points).

```python
# target parameters of f(x) = 4 - 3x - 2x^2 + 3x^3
a = 3
b = -2
c = -3 
d = 4

def f(x) : return a*(x**3) + b*(x**2) + c*x + d

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 5000)
x_valid = np.random.uniform(-1, 1, 500)
x_valid.sort()
y_target = f(x_valid) # ideal (target) function

sigma = 0.50 # noise standard deviation
y_train = np.random.normal(f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(f(x_valid), sigma)

plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

After some tinkering, I came up with the following model:

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))
model.add(Dense(5, activation='exponential'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1, activation='linear'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()

history = model.fit(x=x_train, y=y_train, verbose=False,
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))

# return weights and biases
print()
w = model.get_weights()
#print(w)

# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

This model is maybe a bit overengineered, but it works pretty well **inside the range** with a small loss (evaluated against the exact function).

```python
x_valid_extra = np.random.uniform(-2, 2, 500)
x_valid_extra.sort()
y_target_extra = f(x_valid_extra) # ideal (target) function
y_valid_extra = np.random.normal(f(x_valid_extra), sigma)

# evaluate model
score = model.evaluate(x_valid_extra, y_valid_extra, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

x_predicted_extra = np.random.uniform(-2, 2, 200)
y_predicted_extra = model.predict(x_predicted_extra)
plt.scatter(x_predicted_extra, y_predicted_extra, color='r')
plt.plot(x_valid_extra, y_target_extra)
plt.grid(True)
plt.show()
```

However, it is suffiecient to extend the range of the dataset from [-1, 1] to [-2, 2] to show the limits of this model. The model can only predict the curve inside the training range, while it is absolutely clueless about anything happening on the outside.

<span style="color:gray">Try to extend the model to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

### Solution:

I tried to use the same model structure as before

```python
def f(x, y) : return np.sin(x**2 + y**2)

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1.5, 1.5, 5000)
x_valid = np.random.uniform(-1.5, 1.5, 500)
x_valid.sort()
y_train = np.random.uniform(-1.5, 1.5, 5000)
y_valid = np.random.uniform(-1.5, 1.5, 500)
y_valid.sort()

xy_train = np.stack([x_train, y_train], axis=1)
xy_valid = np.stack([x_valid, y_valid], axis=1)

z_target = f(x_valid, y_valid) # ideal (target) function

sigma = 0.50 # noise standard deviation

z_train = np.random.normal(f(x_train, y_train), sigma) # actual measures from which we want to guess regression parameters
z_valid = np.random.normal(f(x_valid, y_valid), sigma)

inputs = tf.keras.layers.Input(shape=(2,))
l1 = Dense(5, activation="exponential")(inputs)
l2 = Dense(10, activation="tanh")(l1)
outputs = Dense(1, activation="linear")(l2)
model = keras.Model(inputs=inputs, outputs=outputs)

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()

history = model.fit(x=xy_train, y=z_train, verbose=False,
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, z_valid))

# return weights and biases
print()
w = model.get_weights()
#print(w)

# evaluate model
score = model.evaluate(xy_valid, z_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(xy_valid, z_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

x_predicted = np.random.uniform(-1.5, 1.5, 300)
y_predicted = np.random.uniform(-1.5, 1.5, 300)
z_predicted = model.predict(np.stack([x_predicted, y_predicted], axis=1))
#plt.scatter(x_predicted, y_predicted,color='r')
#plt.plot(x_valid, y_target)
#plt.grid(True)
#plt.show()
```
