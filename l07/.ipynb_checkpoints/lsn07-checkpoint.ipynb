Use the NSL simulator code to <span style="color:red">print out instantaneous values of potential energy per particle, $U/N$, along **equilibrated** Monte Carlo simulations</span> (of $M=5\times 10^5$ MC steps and where the Metropolis acceptance ratio has been fixed approximately to 50%) in the following thermodynamic conditions:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

1. <span style="color:red">Once this data has been produced, compute and make pictures with a python script of the autocorrelation functions of $U/N$ to approximately estimate how large the blocks in your simulations should be.
2. Use the same data to study, with a python script, the estimation of the statistical uncertainties of the expectation value of $U/N$ for different size of the blocks in the data blocking technique, from $L=M/N=10$ to $L=5\times 10^3$, being $N$ the number of blocks. <span style="color:red">Make pictures of these statistical uncertainties as a function of $L$.</span> What you observe in the statistical uncertainties after that the block size is large enough (from the point of view of the data blocking technique)?

### Solution:

I simulated 500 thousand Monte Carlo steps for each phase. I fixed the acceptance ratio to approximately 50% with the sole exception of the gasseous phase, for which there was no way of reducing it under 62% using the chosen parameters. I performed the simulations starting from a crystalline configuration:

```python
import matplotlib
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy as np

phases = ["solid", "liquid", "gas"]

calibration = dict()
temporary = dict()
for phase in phases:
    calibration[phase] = pd.read_csv(f'mc_{phase}_calibration/out/potential_energy.dat', delimiter='\t')
    temporary[phase] = calibration[phase].iloc[:1000]
    #print(calibration[phase])

ax = dict()
fig, (ax["solid"], ax["liquid"], ax["gas"]) = plt.subplots(3, 1, figsize=(17,30))
for phase in phases:
    ax[phase].errorbar(calibration[phase]['BLOCK:'], calibration[phase]['ACTUAL_PE:'], yerr=0)
    ax[phase].set_title(phase)
    ax[phase].tick_params(axis='x', labelcolor='b')
    ax[phase].set_xlabel('steps', loc='right', color='b')
    ax[phase].set_ylabel('$U$')
    ax[phase].grid(True)
    aa = ax[phase].twiny()
    aa.errorbar(temporary[phase]['BLOCK:'], temporary[phase]['ACTUAL_PE:'], yerr=0, color='r')
    aa.tick_params(axis='x', labelcolor='r')
    aa.set_xlabel('first 1000 steps', loc='right', color='r')
    #aa.legend()
    #ax[phase].legend(loc='upper center', ncol=3, fancybox=False)
    
plt.show()
```

All phases reach equilibrium well before our usual 1000 steps. I decided nonetheless to cut the first "block" of 1000 steps from the data to perform the autocorrelation analysis:

```python
def expo(x, tau):
    return np.exp(-x/tau)
```

```python
from scipy.optimize import curve_fit

proposed_blocksize=1000

for phase in phases:
    temporary[phase] = calibration[phase].iloc[1000:]

autoc = dict()
for phase in phases:
    autoc[phase] = list()
    for lag in range(proposed_blocksize):
        autoc[phase].append(temporary[phase]['ACTUAL_PE:'].autocorr(lag))

lag = np.arange(proposed_blocksize)
tau = dict()
for phase in phases:
    fit, cov = curve_fit(expo, lag, autoc[phase])
    tau[phase]=fit[0]
    print(phase+' tau:', tau[phase])

ax = dict()
fig, (ax["solid"], ax["liquid"], ax["gas"]) = plt.subplots(3, 1, figsize=(17,30))
for phase in phases:
    ax[phase].plot(autoc[phase], label=phase+' phase autocorrelation')
    ax[phase].plot(lag, expo(lag, tau[phase]), label=r'exponential fit, $\tau$=' + f'{tau[phase]:.2f}')
    ax[phase].set_title(phase)
    ax[phase].set_xlabel('lag')
    ax[phase].set_ylabel('autocorrelation')
    ax[phase].grid(True)
    #ax[phase].legend(loc='upper center', ncol=3, fancybox=False)
    ax[phase].legend(loc='upper right', prop={'size':20})
    
plt.show()
```

In each and every case, though, the autocorrelation function fits an exponential with $\tau \approx 10^1$ (except for the gasseous phase for which it is even smaller), which means that an appropriate blocksize should be at least a few tens of steps long.

I then computed how statistical uncertainties depend on the blocksize. I used blocking average to compute the average value of the potential energy and the relative standard deviation, using blocks of size 10 (the smallest reasonable size according to the results of the autocorrelation fit), 20, 50, 100, 200, 500, 1000, 2000 and 5000.

```python
from collections import OrderedDict

Ls = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]

N, M = dict(), dict()
block, std = dict(), dict()
for phase in phases:
    M[phase] = len(calibration[phase]['BLOCK:'])
    N[phase] = dict()
    block[phase], std[phase] = OrderedDict(), OrderedDict()
    for L in Ls:
        N[phase][L] = int(M[phase] / L)
        #print(N[phase][L])
        block[phase][L] = list()
        for i in range(N[phase][L]):
            block[phase][L].append(calibration[phase]['ACTUAL_PE:'].iloc[i*L:(i+1)*L].mean())
        std[phase][L] = np.std(np.array(block[phase][L]))
        #print(phase, L, std[phase][L])

plt.figure(figsize=(17,10))
plt.plot(Ls, std['solid'].values(), label='solid')
plt.plot(Ls, std['liquid'].values(), label='liquid')
plt.plot(Ls, std['gas'].values(), label='gas')
plt.xlabel('L', loc='right')
plt.ylabel('standard error', loc='top')
#plt.xscale('log')
plt.grid(True)
plt.legend(loc='upper right', prop={'size':20})
plt.show()
```

The statistical uncertainty decreases with the blocksize, but plateaus when $L \approx 10^3$.

By means of your upgraded NSL simulator code, equilibrate and <span style="color:red">perform MC NVT simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:

show in pictures the obtained average values and uncertainties for the potential energy per particle, $U/N$, the pressure $P$ and the radial distribution function $g(r)$ in LJ units ... <span style="color:red"> and compare your MC results for the radial distribution function, $g(r)$, with those obtained with Molecular Dynamics NVE simulations in similar thermodynamic conditions.</span>

### Solution:

Considering the results of the previous exercise, I chose to keep the usual approach of 1000 blocks of size 1000. I let the system equilibrate for 1 block before the start of acquisition:

```python
variables = ["pressure", "potential_energy"]

data = dict()
for var in variables:
    data[var] = dict()
    for phase in phases:
        data[var][phase] = pd.read_csv(f'mc_{phase}/out/{var}.dat', delimiter='\t')

fig, axs = plt.subplots(2, 3, figsize=(17,10))
for ax, variable in zip(axs, variables):
    for a, phase in zip(ax, phases):
        a.errorbar(data[variable][phase]['BLOCK:'], data[variable][phase].iloc[:,2], yerr=data[variable][phase]['ERROR:'])
        a.set_title(phase+': '+variable)
        a.set_xlabel('blocks', loc='right')
        a.set_ylabel(variable, loc='top')
        a.grid(True)
    
plt.show()
```

```python
algos = ["mc", "md"]

gofr = dict()
blocks, bins = dict(), dict()
for algo in algos:
    gofr[algo] = dict()
    blocks[algo], bins[algo] = dict(), dict()
    directory = 'l07'
    if algo == 'md': directory = 'l04'
    for phase in phases:
        gofr[algo][phase] = pd.read_csv(f'../{directory}/{algo}_{phase}/out/gofr.dat', delimiter='\t')
        blocks[algo][phase] = gofr[algo][phase].iloc[-1, 0]
        bins[algo][phase] = int(gofr[algo][phase].shape[0] / blocks[algo][phase])
        gofr[algo][phase] = gofr[algo][phase].iloc[-bins[algo][phase]:] #Get only the last block worth of data
        #print(gofr[algo][phase])
        #print(algo, phase, gofr[algo][phase])
        
fig, axs = plt.subplots(1, 3, figsize=(17,4))
for a, phase in zip(axs, phases):
    a.errorbar(gofr["mc"][phase]['DISTANCE:'], gofr["mc"][phase]['AVE_GOFR:'], yerr=gofr["mc"][phase]['ERROR:'], label='MC')
    a.errorbar(gofr["md"][phase]['DISTANCE:'], gofr["md"][phase]['AVE_GOFR:'], yerr=gofr["md"][phase]['ERROR:'], color='r', label='MD')
    a.set_title(algo+': '+phase)
    a.set_xlabel('r', loc='right')
    a.set_ylabel('g(r)', loc='top')
    a.legend()
    a.grid(True)
```

The $g(r)$ functions, bar some small differences, are the same for the Molecular Dynamics NVE simulation and for the Monte Carlo NVT simulation.
