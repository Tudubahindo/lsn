Use the NSL simulator code to <span style="color:red">print out instantaneous values of potential energy per particle, $U/N$, along **equilibrated** Monte Carlo simulations</span> (of $M=5\times 10^5$ MC steps and where the Metropolis acceptance ratio has been fixed approximately to 50%) in the following thermodynamic conditions:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

### Solution:

```python
import matplotlib
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy as np

phases = ["solid", "liquid", "gas"]

calibration = dict()
for phase in phases:
    calibration[phase] = pd.read_csv(f'mc_{phase}_calibration/out/potential_energy.dat', delimiter='\t')
    #print(calibration[phase])

ax = dict()
fig, (ax["solid"], ax["liquid"], ax["gas"]) = plt.subplots(3, 1, figsize=(17,30))
for phase in phases:
    ax[phase].errorbar(calibration[phase]['BLOCK:'], calibration[phase]['ACTUAL_PE:'], yerr=0)
    ax[phase].set_title(phase)
    ax[phase].set_xlabel('steps')
    ax[phase].set_ylabel('$U$')
    ax[phase].grid(True)
    #ax[phase].legend(loc='upper center', ncol=3, fancybox=False)
    
plt.show()
```

```python
def expo(x, tau):
    return np.exp(-x/tau)
```

```python
from scipy.optimize import curve_fit

proposed_blocksize=1000

autoc = dict()
for phase in phases:
    autoc[phase] = list()
    for lag in range(proposed_blocksize):
        autoc[phase].append(calibration[phase]['ACTUAL_PE:'].autocorr(lag))

lag = np.arange(proposed_blocksize)
tau = dict()
for phase in phases:
    fit, cov = curve_fit(expo, lag, autoc[phase])
    tau[phase]=fit[0]
    print(phase+' tau:', tau[phase])

ax = dict()
fig, (ax["solid"], ax["liquid"], ax["gas"]) = plt.subplots(3, 1, figsize=(17,30))
for phase in phases:
    ax[phase].plot(autoc[phase], label=phase+' phase autocorrelation')
    ax[phase].plot(lag, expo(lag, tau[phase]), label=r'exponential fit, $\tau$=' + f'{tau[phase]:.2f}')
    ax[phase].set_title(phase)
    ax[phase].set_xlabel('lag')
    ax[phase].set_ylabel('autocorrelation')
    ax[phase].grid(True)
    #ax[phase].legend(loc='upper center', ncol=3, fancybox=False)
    ax[phase].legend(loc='upper right', prop={'size':20})
    
plt.show()
```

```python
from collections import OrderedDict

Ls = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]

N, M = dict(), dict()
block, std = dict(), dict()
for phase in phases:
    M[phase] = len(calibration[phase]['BLOCK:'])
    N[phase] = dict()
    block[phase], std[phase] = OrderedDict(), OrderedDict()
    for L in Ls:
        N[phase][L] = int(M[phase] / L)
        #print(N[phase][L])
        block[phase][L] = list()
        for i in range(N[phase][L]):
            block[phase][L].append(calibration[phase]['ACTUAL_PE:'].iloc[i*L:(i+1)*L].mean())
        std[phase][L] = np.std(np.array(block[phase][L]))
        #print(phase, L, std[phase][L])

plt.figure(figsize=(17,10))
plt.plot(Ls, std['solid'].values(), label='solid')
plt.plot(Ls, std['liquid'].values(), label='liquid')
plt.plot(Ls, std['gas'].values(), label='gas')
plt.xlabel('L', loc='right')
plt.ylabel('standard error', loc='top')
#plt.xscale('log')
plt.grid(True)
plt.legend(loc='upper right', prop={'size':20})
plt.show()
```

### Solution:

```python
variables = ["pressure", "potential_energy"]

data = dict()
for var in variables:
    data[var] = dict()
    for phase in phases:
        data[var][phase] = pd.read_csv(f'mc_{phase}/out/{var}.dat', delimiter='\t')

fig, axs = plt.subplots(2, 3, figsize=(17,10))
for ax, variable in zip(axs, variables):
    for a, phase in zip(ax, phases):
        a.errorbar(data[variable][phase]['BLOCK:'], data[variable][phase].iloc[:,2], yerr=data[variable][phase]['ERROR:'])
        a.set_title(phase+': '+variable)
        a.set_xlabel('blocks', loc='right')
        a.set_ylabel(variable, loc='top')
        a.grid(True)
    
plt.show()
```

```python
algos = ["mc", "md"]

gofr = dict()
blocks, bins = dict(), dict()
for algo in algos:
    gofr[algo] = dict()
    blocks[algo], bins[algo] = dict(), dict()
    directory = 'l07'
    if algo == 'md': directory = 'l04'
    for phase in phases:
        gofr[algo][phase] = pd.read_csv(f'../{directory}/{algo}_{phase}/out/gofr.dat', delimiter='\t')
        blocks[algo][phase] = gofr[algo][phase].iloc[-1, 0]
        bins[algo][phase] = int(gofr[algo][phase].shape[0] / blocks[algo][phase])
        gofr[algo][phase] = gofr[algo][phase].iloc[-bins[algo][phase]:] #Get only the last block worth of data
        #print(gofr[algo][phase])
        #print(algo, phase, gofr[algo][phase])
        
fig, axs = plt.subplots(2, 3, figsize=(17,10))
for ax, algo in zip(axs, algos):
    for a, phase in zip(ax, phases):
        a.errorbar(gofr[algo][phase]['DISTANCE:'], gofr[algo][phase]['AVE_GOFR:'], yerr=gofr[algo][phase]['ERROR:'])
        a.set_title(algo+': '+phase)
        a.set_xlabel('r', loc='right')
        a.set_ylabel('g(r)', loc='top')
        a.grid(True)
```
